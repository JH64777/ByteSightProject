```python
'''
   (2024 08 04)
Train 700
Validation 300

-      
-  hex   + (   -    )  
'''

def ReadyXdata(path): #   
    Xdata = []
    files = os.listdir(path)
    processedCount = 0 #    (    )

    for file in files:
        with open(path + '/' + file, 'rb') as f:
            hdata = list(f.read())
            count, length, patches = plt.hist(hdata, bins=256, range=(0, 255)) #   
            plt.clf()

            fileSize = os.path.getsize(path + '/' + file) #  
            img = cv2.imread(path + '/' + file) #       
            count = np.append(count, abs(fileSize - len(img) * len(img[0])) / 1024) # (  -    ) 
            count = np.log1p(count) #   (   ,      )

            Xdata.append(count) #   
            processedCount += 1
            print(f"{file} is well done! : count {processedCount}")
    
    return np.array(Xdata)
```

```python
trainPlainX = ReadyXdata("E:/pythonStegano/AI_workbench/finalData/Train/plain")
trainStegoX = ReadyXdata("E:/pythonStegano/AI_workbench/finalData/Train/stego")

trainSet = np.concatenate((trainPlainX, trainStegoX))
```

```python
trainY = np.concatenate((np.zeros(trainPlainX.shape[0]), np.ones(trainStegoX.shape[0])))
print("WellDone!")
```

```python
validPlainX = ReadyXdata("E:/pythonStegano/AI_workbench/finalData/Validation/plain")
validStegoX = ReadyXdata("E:/pythonStegano/AI_workbench/finalData/Validation/stego")
validSet = np.concatenate((validPlainX, validStegoX))
print("WellDone!")
```

```python
validY = np.concatenate((np.zeros(validPlainX.shape[0]), np.ones(validStegoX.shape[0])))

print(validSet.shape)
print(validY.shape)
```

```python
#    loss: 0.4190 - accuracy: 0.8236 - val_loss: 0.6301 - val_accuracy: 0.7100
l2_reg = tf.keras.regularizers.l2(0.4)
# l1_reg = tf.keras.regularizers.l1(0.03)

X = tf.keras.layers.Input(shape=[257])
H = tf.keras.layers.Dense(60, activation='leaky_relu')(X)
H = tf.keras.layers.Dropout(0.3)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Dense(30, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
H = tf.keras.layers.Dropout(0.2)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Dense(20, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
H = tf.keras.layers.Dropout(0.1)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Dense(10, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
H = tf.keras.layers.Dropout(0.1)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Dense(5, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
Y = tf.keras.layers.Dense(1, activation="sigmoid")(H)
model = tf.keras.models.Model(X, Y)

# X = tf.keras.layers.Input(shape=[257])
# H = tf.keras.layers.Dense(50, activation='leaky_relu', kernel_regularizer=l2_reg)(X)
# H = tf.keras.layers.Dropout(0.2)(H)
# H = tf.keras.layers.BatchNormalization()(H)
# H = tf.keras.layers.Dense(20, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
# H = tf.keras.layers.Dropout(0.2)(H)
# H = tf.keras.layers.BatchNormalization()(H)
# H = tf.keras.layers.Dense(10, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
# H = tf.keras.layers.Dense(5, activation='leaky_relu', kernel_regularizer=l2_reg)(H)
# Y = tf.keras.layers.Dense(1, activation="sigmoid")(H)
# model = tf.keras.models.Model(X, Y)

Adam = tf.keras.optimizers.Adam(learning_rate=0.0001)
# AdamWOp = tf.keras.optimizers.AdamW(learning_rate=0.0001, weight_decay=0.01)
model.compile(loss="binary_crossentropy", optimizer=Adam, metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(trainSet, trainY, epochs=300, validation_data=(validSet, validY))
```
